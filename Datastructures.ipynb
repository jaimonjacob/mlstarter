# Data structures

### Lists: Ordered, mutable, allows duplicate values




Python list methods (Some common for Tuple as well) https://www.programiz.com/python-programming/methods/list/append

```python
# Sort - Changes list
list_one = [3, 8, 5, 6]
list_one.sort()
print(f'list_one: {list_one}')

# Sorted - No changes to original list
list_two = [10, 4, 98, 62]
newlist_two = sorted(list_two)
print(f'list_two: {list_two}')
print(f'new list_two: {newlist_two}')
```

```python
# Create a new list with same items
list_one = [1] * 5
print(f'list_one: {list_one}')

# Concatenate two lists
list_two = [2] * 5
print(f'list_one concatenated wtih list_two: {list_one + list_two}')
```

```python
# Slicing a list
list_one = [1, 2, 3, 4, 5, 6, 7, 8]
list_two = list_one[0:3]
list_three = list_one[:3]
list_four = list_one[1:]

# Using step to take every step items (here 2)
list_five = list_one[: : 2]

print(f'list_one sliced to create list two: {list_two}')
print(f'list_one sliced to create list three wtih specifying the last  paramter: {list_three}')
print(f'list_one sliced to create list four wtih specifying the first  paramter: {list_four}')
print(f'list_one sliced to create list five wtih specifying the step  paramter: {list_five}')
```

```python
# Copying lists
list_one = [1, 2, 3, 4, 5]

# This method creates a connected copy. Items changed in one copy is changed in the copy as well
list_two = list_one
list_one.insert(2, 'a')
print(f'list one: {list_one}')
print(f'list_two: {list_two}')

# This method creates a complete copy. Items changed in one copy is not changed in the copy
list_three = list_one.copy()
list_one.insert(3, 'b')
print(f'list_one: {list_one}')
print(f'list_three: {list_three}')

# THe other ways to make a complete copy
list_four = list(list_one)
list_five = list_one[:]
```

```python
# List comprehension  - Create new lists by using expressions within one list (similar to map in JS)
list_one = [1, 2, 3, 4, 5]
list_two = [i+10 for i in list_one]
print(f'list two: {list_two}')
```

```python
# Desctructuring/unpacking lists in Python

list_one = ['john', 'mary', 'Lia']
first, second, third = list_one
print(f'unpacking single items: {first, second, third}')


list_two = ['john', 'mary', 'Lia', 'jango', 'rambo']
person_one, *persons, person_three = list_two
print(f'unpacking group items: {person_one, persons, person_three}')
```

## Tuple: Ordered, immutable, allows duplicate values

```python
# Creating tuple from values
tuple_one = ('Max', 12, True)

# Creating tuple from an iterable
tuple_two = tuple(['abc', 12, False])


print(f'my_type: {tuple_one}')
print(f'my_type: {tuple_two}')
```

```python
# Extracting item
tuple_one = ('Max', 12, True)
print(tuple_one[2])

# Assignment is not possible
# tuple_one[2] = False
```

```python
# Efficiency is using Tuple whenever possible
import sys
import timeit

list_one = []
for value in range (1, 1000):
  list_one.append(value)

tuple_one = tuple(list_one)

# The tuple is much smaller in size so preferred for usage
print(f'Size of list: {sys.getsizeof(list_one)} bites')
print(f'Size of tuple: {sys.getsizeof(tuple_one)} bites')

# The tuple also is much faster to create so preferred for usage
print(f'''Time to create the list 10000000 times: {timeit.timeit(stmt='[1, 2, 3, 4, 5]', number=10000000)}''')
print(f'''Time to create the tuple 10000000 times: {timeit.timeit(stmt='(1, 2, 3, 4, 5)', number=10000000)}''')
```

## Dictionary: Key-value pairs, unordered, mutable

Python dictionary methods: https://www.programiz.com/python-programming/methods/dictionary/clear

```python
# Creating a dicionary
dict_one = dict(name = 'joe', age= 12, married=True)

# Extracting value
dict_one['name']

# Adding value
dict_one['location'] = 'USA'

# Deleting value
del dict_one['married']

# Iterating value
for key, value in dict_one.items():
  print(f'The key and values are {key, value}')

# Making an actual copy
dict_two = dict_one.copy()
#or
dict_two = dict(dict_one)

# Merging two dictionaries
dict_three = dict(name = 'mary', age= 14, married=False, location = 'USA')
dict_two.update(dict_three)

print(f'Dict one: {dict_one}')
print(f'Dict one: {dict_two}')
```
