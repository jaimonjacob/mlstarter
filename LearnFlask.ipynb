### Database configuration

```python
app = Flask(__name__,
            static_url_path='', 
            static_folder='static',
            template_folder='templates'
            )

app.config['SECRET_KEY'] = 'secret key'

# SQLite User database configuraiton
# create the db using - sqlite3 users.db "" followed by .databases and .quit
# Then run in the pythonshell
# from app import db
# db.create()

app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///users.db' 
   
# MYSQL db configuraiton
# User either create_db.py or xamp to create the database beore running db.create_all() from python shell
app.config['SQLALCHEMY_DATABASE_URI'] = 'mysql+pymysql://root:password123@localhost/users'


# Create the database using sql alchemy
db = SQLAlchemy(app)

# model for user db table
class Users(db.Model):
     id = db.Column(db.Integer, primary_key=True)
     name = db.Column(db.String(200), nullable=False)
     email = db.Column(db.String(120), nullable=False, unique=True)
     date_added = db.Column(db.DateTime, default=datetime.utcnow)

     def __repr__(self):
          return '<Name %r>' %self.name
    

#create_DB
import mysql.connector

mydb = mysql.connector.connect(host='localhost', user='root', passwd='password123')

my_cursor = mydb.cursor()

my_cursor.execute("CREATE DATABASE users")
my_cursor.execute("SHOW DATABASES")

for db in my_cursor:
    print(db)    
```
